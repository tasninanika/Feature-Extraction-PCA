```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn import datasets
```

```python
from sklearn.decomposition import PCA
```

```python
df = datasets.load_iris()
```

```python
df.feature_names
```

```python
print(df.DESCR)
```

```python
X = df.data
X
```

```python
X = pd.DataFrame(df.data, columns = df.feature_names) #convert to dataframe
X
```

```python
feature_names = X.columns.tolist()  # Get actual feature names
feature_names
```

```python
# Step 1: Standardize the dataset
scaler = StandardScaler()            
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# Step 2: Compute the covariance matrix
cov_matrix = np.cov(X_scaled.T)
cov_matrix
```

```python
# Step 3: Compute Eigenvalues & Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
eigenvalues, eigenvectors
```

```python
# Step 4: Sort eigenvalues and corresponding eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]  # Sort in descending order
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
eigenvalues
eigenvectors
```

```python
# Step 5: Calculate explained variance percentage
explained_variance_ratio = eigenvalues / np.sum(eigenvalues) * 100
print("explained_variance_ratio ", explained_variance_ratio )
cumulative_variance = np.cumsum(explained_variance_ratio)
print("cumulative_variance", cumulative_variance)
```

```python
# Step 6: Select the number of principal components to keep 95% variance
num_components = np.argmax(cumulative_variance >= 80) + 1  # First index reaching 95% variance
print("Number of Principal Components selected:", num_components)
```

```python
# Step 7: Project data onto new principal components
X_pca = np.dot(X_scaled, eigenvectors[:, :num_components])
X_pca
```

```python
# Step 8: Visualize the explained variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_variance, marker='o', linestyle='--')
plt.axhline(y=80, color='r', linestyle='--', label='80% Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('Explained Variance vs. Number of Components')
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 9: Find Top Contributing Features for PCA
selected_features = pd.DataFrame(eigenvectors[:, :num_components], index=feature_names)
top_features = selected_features.abs().sum(axis=1).nlargest(num_components).index.tolist()

print("\nTop Contributing Features for PCA:", top_features)
```
