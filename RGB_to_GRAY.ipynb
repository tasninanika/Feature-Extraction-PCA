```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# load image
img_bgr = cv2.imread("/content/drive/MyDrive/DIP/dog2.jpg")
```

```python
# convert the bgr color to rgb
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
# convert the rgb color to gray
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(10,5))

# Default BGR Image
plt.subplot(1,3,1)
plt.imshow(img_bgr)
plt.title("BGR")
plt.axis("off")

# Converted RGB Image
plt.subplot(1,3,2)
plt.imshow(img_rgb)
plt.title("RGB")
plt.axis("off")

# Converted Gray image
plt.subplot(1,3,3)
plt.imshow(img_gray, cmap='gray')
plt.title("Gray")
plt.axis("off")

plt.show()
```
