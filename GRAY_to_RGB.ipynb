```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# load image default bgr
img_bgr = cv2.imread("/content/drive/MyDrive/DIP/dog3.jpg")
print("BGR image shape:", img_bgr.shape)
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
print("RGB image shape:", img_rgb.shape)
```

```python
# convert bgr to gray
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_RGB2GRAY)
print("Grayscale image shape:", img_gray.shape)
```

```python
# convert gray to rgb
gray_to_rgb = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
print("Gray to RGB image shape:", gray_to_rgb.shape)
```

```python
plt.figure(figsize=(10,5))

# Default BGR Image
plt.subplot(1,4,1)
plt.imshow(img_bgr)
plt.title("BGR")
plt.axis("off")

# RGB
plt.subplot(1,4,2)
plt.imshow(img_rgb)
plt.title("RGB")
plt.axis("off")

# Grayscale
plt.subplot(1,4,3)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

# Gray â†’ RGB
plt.subplot(1,4,4)
plt.imshow(gray_to_rgb)
plt.title("Gray to RGB")
plt.axis("off")

plt.show()
```
