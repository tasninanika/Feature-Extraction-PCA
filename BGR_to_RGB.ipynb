```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# image load
img_bgr = cv2.imread("/content/drive/MyDrive/DIP/dog.jpg")
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,5))

# Default BGR Image
plt.subplot(1,2,1)
plt.imshow(img_bgr)
plt.title("BGR (default)")
plt.axis("off")

# Converted RGB Image
plt.subplot(1,2,2)
plt.imshow(img_rgb)
plt.title("RGB (Correct Colors)")
plt.axis("off")

plt.show()
```
